```python
### HISTOGRAM OF GRADES
```

```python
plt.hist(df['G1'],bins=10,alpha=0.5,label='G1')
plt.hist(df['G2'],bins=10,alpha=0.5,label='G2')
plt.hist(df['G3'],bins=10,alpha=0.5,label='G3')
plt.legend()
plt.xlabel=('grades')
plt.ylabel=( 'students')
plt.title('grades histrogram')
plt.show()
```

### SCATTER PLOT OF STUDYTIME AND GRADES

```python
plt.scatter(df['studytime'],df['G1'],label='G1')
plt.scatter(df['studytime'],df['G2'],label='G2')
plt.scatter(df['studytime'],df['G3'],label='G3')
plt.xlabel=('studytime')
plt.ylabel=('grades')
plt.title('studytime vs grades')
plt.legend()
plt.show()
```

```python
### BAR CHART OF MALE AND FEMALE AVERAGE SCORE 
```

```python
avg_scores = df.groupby('sex')[['G1','G2','G3']].mean()
avg_scores.plot(kind='bar')
plt.xlabel = ('gender')
plt.ylabel = ('score')
plt.title('male vs female average score ')
plt.show()
```

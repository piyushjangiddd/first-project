```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\Users\MAIN POINT\Downloads\archive\Titanic-Dataset.csv')
```

```python
print(df)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df['Age'].isnull().sum()
```

```python
df.groupby('Sex')['Survived'].sum()
```

```python
df.groupby('Pclass')['Survived'].mean()
```

```python
df.groupby('Age')['Survived'].mean()
```

```python
bins = [0,12,18,35,60,80,100]
labels=['child','teen','young','adult','senior','super senior']
df['AgeGroup']=pd.cut(df['Age'],bins=bins,labels=labels)
```

```python
df.groupby('AgeGroup')['Survived'].sum()
```

```python
df.groupby('AgeGroup')['Survived'].mean()
```

```python
df.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.xlabel=('Gender')
plt.ylabel=('Survival Rate')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
df.groupby('Pclass')['Survived'].mean().plot(kind='bar')
plt.xlabel=('Pclass')
plt.ylabel=('Survival Rate')
plt.title('Survival Rate by Pclass')
plt.show()
```

```python
df['Age'].plot(kind='hist',bins=20)
plt.xlabel=('Age')
plt.ylabel=('Frequency')
plt.title('Histogram of passenger ages')
plt.show()
```
